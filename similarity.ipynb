```python
import pandas as pd
import numpy as np

# Define the user-item matrix
data = {
    "P1": [3, np.nan, 5, np.nan, np.nan],
    "P2": [3, 2, 4, 4, 3],
    "P3": [1, 2, np.nan, 5, 5],
    "P4": [3, 5, 4, 4, 3],
    "P5": [np.nan, np.nan, 2, 4, 3]
}
index = ["User1", "User2", "User3", "User4", "User5"]
df = pd.DataFrame(data, index=index)

# Cosine similarity function between two users
def cosine_similarity(user1, user2):
    ratings_user1 = df.loc[user1]
    ratings_user2 = df.loc[user2]
    common_ratings = ratings_user1.notna() & ratings_user2.notna()
    
    if not common_ratings.any():
        return 0
    
    ratings_user1 = ratings_user1[common_ratings]
    ratings_user2 = ratings_user2[common_ratings]
    
    numerator = np.dot(ratings_user1, ratings_user2)
    denominator = np.sqrt(np.sum(ratings_user1 ** 2)) * np.sqrt(np.sum(ratings_user2 ** 2))
    
    return numerator / denominator if denominator != 0 else 0

# Calculate similarity between User1 and each other user
target_user = "User1"
similarities = {}

for user in df.index:
    if user != target_user:
        similarity = cosine_similarity(target_user, user)
        similarities[user] = similarity

# Display the similarities
for user, sim in similarities.items():
    print(f"Cosine similarity between {target_user} and {user}: {sim}")
```

```python
import pandas as pd
import numpy as np

# Define the user-item matrix
data = {
    "P1": [3, np.nan, 5, np.nan, np.nan],
    "P2": [3, 2, 4, 4, 3],
    "P3": [1, 2, np.nan, 5, 5],
    "P4": [3, 5, 4, 4, 3],
    "P5": [np.nan, np.nan, 2, 4, 3]
}
index = ["User1", "User2", "User3", "User4", "User5"]
df = pd.DataFrame(data, index=index)

# Function to calculate the mean rating for a user
def mean_rating(user):
    return df.loc[user].mean()

# Cosine similarity function between two users
def cosine_similarity(user1, user2):
    ratings_user1 = df.loc[user1]
    ratings_user2 = df.loc[user2]
    common_ratings = ratings_user1.notna() & ratings_user2.notna()
    
    if not common_ratings.any():
        return 0
    
    ratings_user1 = ratings_user1[common_ratings]
    ratings_user2 = ratings_user2[common_ratings]
    
    numerator = np.dot(ratings_user1, ratings_user2)
    denominator = np.sqrt(np.sum(ratings_user1 ** 2)) * np.sqrt(np.sum(ratings_user2 ** 2))
    
    return numerator / denominator if denominator != 0 else 0

# Calculate similarity between User1 and each other user
target_user = "User1"
similarities = {}

for user in df.index:
    if user != target_user:
        similarity = cosine_similarity(target_user, user)
        similarities[user] = similarity

# Predict rating for Product (P5) for User1 using the adjusted equation
def predict_rating_adjusted(user, product, top_n=2):
    # Mean rating of the target user
    mean_target_user = mean_rating(user)

    # Sort the similarities in descending order and take the top N
    sorted_similarities = sorted(similarities.items(), key=lambda item: item[1], reverse=True)[:top_n]
    
    numerator = 0
    denominator = 0
    for other_user, sim in sorted_similarities:
        # Only consider ratings for the product from other users
        if not np.isnan(df.loc[other_user, product]):
            mean_other_user = mean_rating(other_user)
            numerator += sim * (df.loc[other_user, product] - mean_other_user)
            denominator += sim
            
            # Debugging output for understanding the calculation
            print(f"Similarity with {other_user}: {sim}, Rating: {df.loc[other_user, product]}, "
                  f"Mean Rating: {mean_other_user}, Contribution: {sim * (df.loc[other_user, product] - mean_other_user)}")
    
    # Calculate predicted rating (avoid division by zero)
    if denominator != 0:
        predicted_rating = mean_target_user + (numerator / denominator)
    else:
        predicted_rating = np.nan
    
    return predicted_rating

# Predict rating for User1 on Product 5 using the adjusted formula
predicted_rating_adjusted = predict_rating_adjusted(target_user, "P5", top_n=2)
print(f"pearson Predicted rating for {target_user} on Product 5 (adjusted): {predicted_rating_adjusted}")
```

```python
import pandas as pd
import numpy as np

# Define the user-item matrix
data = {
    "P1": [3.0, np.nan, 5.0, np.nan, np.nan],
    "P2": [3.0, 2.0, 4.0, 4.0, 3.0],
    "P3": [1.0, 2.0, np.nan, 5.0, 5.0],
    "P4": [3.0, 5.0, 4.0, 4.0, 3.0],
    "P5": [np.nan, np.nan, 2.0, 4.0, 3.0]
}
index = ["User1", "User2", "User3", "User4", "User5"]
df = pd.DataFrame(data, index=index)

# Pre-compute the mean ratings for each user
mean_ratings = df.mean(axis=1)

# Adjusted Cosine Similarity function between two products
def adjusted_similarity(product1, product2):
    # Get ratings for the two products
    r_1 = df[product1]
    r_2 = df[product2]
    
    # Identify common users who rated both products
    common_users = r_1.notna() & r_2.notna()
    
    if not common_users.any():
        return 0
    
    # Ratings of common users for each product
    r_1_common = r_1[common_users]
    r_2_common = r_2[common_users]
    
    # Use individual mean ratings for each common user
    mean_common_users = mean_ratings[common_users]
    
    # Calculate the numerator
    numerator = np.sum((r_1_common - mean_common_users) * (r_2_common - mean_common_users))
    
    # Calculate the denominator
    denominator_1 = np.sqrt(np.sum((r_1_common - mean_common_users) ** 2))
    denominator_2 = np.sqrt(np.sum((r_2_common - mean_common_users) ** 2))
    
    # Return adjusted cosine similarity
    return numerator / (denominator_1 * denominator_2) if (denominator_1 * denominator_2) != 0 else 0

# Example of calculating adjusted cosine similarity between products
adjusted_similarity_p1_p2 = adjusted_similarity("P1", "P2")
print(f"Adjusted cosine similarity between P1 and P2: {adjusted_similarity_p1_p2}")

# You can also compute the adjusted similarities for other pairs of products
adjusted_similarity_p1_p3 = adjusted_similarity("P1", "P3")
print(f"Adjusted cosine similarity between P1 and P3: {adjusted_similarity_p1_p3}")

adjusted_similarity_p1_p4 = adjusted_similarity("P1", "P4")
print(f"Adjusted cosine similarity between P1 and P4: {adjusted_similarity_p1_p4}")

adjusted_similarity_p1_p5 = adjusted_similarity("P1", "P5")
print(f"Adjusted cosine similarity between P1 and P5: {adjusted_similarity_p1_p5}")

# Function to predict rating for a given user and product using the top 2 highest similarities
def predict_rating_top2(user, product):
    if product in df.columns:
        # Get ratings from the specified user
        user_ratings = df.loc[user]
        
        # Identify rated products
        rated_products = user_ratings.notna()
        
        # Check if the product has not been rated by the user
        if not rated_products[product]:  # User hasn't rated the product yet
            similarities = []
            
            # Collect similarities and ratings for rated products
            for other_product in df.columns:
                if rated_products[other_product]:  # User has rated this product
                    similarity = adjusted_similarity(other_product, product)
                    rating = user_ratings[other_product]
                    if similarity != 0:  # Only consider non-zero similarities
                        similarities.append((similarity, rating))
            
            # Sort similarities by absolute value and get the top 2
            top_similarities = sorted(similarities, key=lambda x: abs(x[0]), reverse=True)[:2]
            
            # Calculate the numerator and denominator
            numerator = sum(sim * rating for sim, rating in top_similarities)
            denominator = sum(abs(sim) for sim, _ in top_similarities)
            
            # Return the predicted rating
            return (numerator / denominator) if denominator != 0 else np.nan
    
    return np.nan  # Return NaN if the product doesn't exist

# Example of predicting a rating for User1 on product P5 using top 2 similarities
predicted_rating_user1_p5_top2 = predict_rating_top2("User1", "P5")
print(f"Predicted rating for User1 on P5 (using top 2 similarities): {predicted_rating_user1_p5_top2:.2f}")

# You can also test for other users and products
predicted_rating_user2_p1_top2 = predict_rating_top2("User2", "P1")
print(f"Predicted rating for User2 on P1 (using top 2 similarities): {predicted_rating_user2_p1_top2:.2f}")
```
